```python
Población = pd.read_csv("https://docs.google.com/spreadsheets/d/13LJWB7oX2rya6n0zeAiw4BQPZ3M4dsb9/export?format=csv")
```

```python
Población[["Provincia", "Hombres", "Mujeres"]]
```

Aqui se puede observar una tabla con la población de Castilla y Leon segun el último censo de 2021. Como se puede observar, en todas las ciudades las mujeres superan en número al hombre, lo que concuerda con la subida de porcentaje de mujeres en España, el cual supera el 50% ya.

```python
Población["Población Total por Provincia"] = Población["Hombres"] + Población["Mujeres"]
```

```python
Población[["Provincia", "Hombres", "Mujeres", "Población Total por Provincia"]]
```

En esta tabla se puede observar además el número total de inhabitantes por municipio, y como se puede observar, Valladolid es el más habitado, con Segovia en último puesto.

```python
Población[(Población ["Población Total por Provincia"] >= 100.000)][["Provincia", "Hombres", "Mujeres", "Población Total por Provincia"]]
```

Aqui se puede observar la cantidad de municipios que superan los 100.000 habitantes, siendo solo 4 de los 9 que hay en Castilla y León, pero no es nada comparado con los 3 millones de Madrid.

```python
Población["Ratio de hombres por cada mujer"] = Población["Hombres"] / (Población["Mujeres"])
```

```python
Población
```

Aqui hay una tabla que concentra los datos más principales: el municipio con el número de mujeres, hombres, la población total y el ratio de hombres por cada mujer, que en ningun caso supera la igualdad numerica, pero se mantiene en torno a los 9 hombres por cada 10 mujeres.

```python
Población.sample(1)
```

Y aqui hay una tabla de una sola fila que muestra los datos de un municipio aleatorio

```python
Población.plot.scatter(y="Provincia", x= "Mujeres")
```

Aqui tenemos una grafica de puntos con las provincias y las mujeres (en miles) que viven en cada una, con lo que se puede observar la gran diferencia entre el primero (Valladolid) y el segundo (Burgos)

```python
GrHombres = pd.DataFrame({'Hombres': [55.384 , 36.362, 82.920, 139.390]}, index=['Palencia', 'Leon', 'Burgos', 'Valladolid'])
display(GrHombres)

Gráfica = GrHombres.plot.pie(y='Hombres', figsize=(6, 5), autopct='%1.1f%%')
```

Aqui tenemos un gráfico circular con el porcentaje de hombres en cuatro provincias: Palencia, Leon, Burgos y Valladolid, y se puede ver como el primero casi dobla al segundo puesto.

```python
Población.sort_values("Ratio de hombres por cada mujer").plot.line(x="Provincia", y="Ratio de hombres por cada mujer") 
```

Aqui obserbamos una gráfica ascendente con el ratio de hombres por cada mujer, en el que se observa a León en último lugar, mientras que Ávila se encuentra primero, con una diferencia de un hombre más por cada 10 mujeres

**Gracias a estos datos podemos observar el actual estado demográfico Español usando como ejemplo Castilla y Leon: Un avance progresivo y acelerado de las mujeres no solo en el ámbito social, sino también en el demográfico. También se puede observar como se supera el millón de habitantes por muy poco, pero con un esparcimiento de la población muy desigual, siendo que entre Burgos y Valladolid ya hay casi un 70% de la población.
